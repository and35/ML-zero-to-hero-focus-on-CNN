Consider the following sets of numbers. Can you see the relationship between them?

x = [-1, 0, 1, 2, 3, 4]   
y = [-2, 1, 4, 7, 10, 13]

As you look at them, you might notice that the value of X is increasing by 1 as you read left to right and the corresponding value of Y is increasing by 3. You probably think that Y equals 3X plus or minus something. Then, you'd probably look at the 0 on X and see that Y is 1, and you'd come up with the relationship Y=3X+1.

That's almost exactly how you would use code to train a model to spot the patterns in the data!

Now, look at the code to do it.

How would you train a neural network to do the equivalent task? Using data! By feeding it with a set of X's and a set of Y's, it should be able to figure out the relationship between them.

```python
 # 0 load the libraries needed 
import tensorflow as tf 
from tensorflow import keras 
import numpy as np
# 1 loading the data
# 1.2 loading data in variables
# 2 sanity check and visualization of dataset 
# 3 normalization: it is not necesary 
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-2.0, 1.0, 4.0, 7.0, 10.0, 13.0], dtype=float)

```

```python
"""definimos el modelo, una NN. es la NN mas sencilla ya que es solo una neurona 
 keras.Sequential() = crear una NN secuencial
[keras.layers.Dense(units=1, input_shape=[1])] = crear una capa, con una neurona (unit), y le pasamos una sola variable de entrada (input_shape )
"""
# 4 desingning the model (solo una neurona)
model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
# compilamos el modelo 
""" 
este paso consiste en asignar una funcion de coste y una algoritmo de correccion 
optimizer="sdg" == es el que mejora la aproximacion, en base a la funcion tenemos que corregir los pesos de la neurona, 
de esto se encarga el optimizador, es como el gradient decent. stochastic gradient descent (sgd) 
loss= "mean_square_error" == una funcion de coste para evaluar que tambien hace el trabajo el algoritmo
"""
# 5 compiling the model 
model.compile(optimizer="sgd", loss= "mean_squared_error")
# ajustamos los parametros, o entrenemos el modelo
# 6 trainig the model 
model.fit(xs, ys, epochs=500)

# 7 evaluating the model 
print(model.predict([10.0]))
# the result is not 31 cause neural networks deal with probabilities
```

The collection of a network, loss function, and optimizer handles the process of guessing the relationship between the numbers, measuring how well they did, and then generating new parameters for new guesses.
